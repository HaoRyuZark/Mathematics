### LÖSUNG AUFGABE 1

#### a) 

Die genaue definition vom Konvergenz lautet, dass sei $\{a_n\}_{n = 1}^{\infty}$ eine Folge mit dem Grenzwert $L$

$$
    \forall \varepsilon > 0, \exists n, N \in \mathbb{N}: \forall n \ge N, |a_n - L| < \varepsilon
$$

Dies bedeutet, dass fuer jedes Epsilon groesser als 0 wir immer eine natuerliche Zahl N so, dass die Differenz aller Folgenglieder von dem Grenzwert kleiner ist als Epsilon. 

#### b)

**Gemeinsamkeiten:**
- Beide beschreiben das Annäherungsverhalten von Folgen.
- Jeder Grenzwert ist auch ein Häufungswert.

**Unterschiede:**
- **Grenzwert:** Höchstens einer; alle Folgenglieder nähern sich diesem Wert an.  
  Formal: $\forall \varepsilon > 0\, \exists N : n \ge N \Rightarrow |a_n - L| < \varepsilon $.

- **Häufungswert:** Es können mehrere existieren; unendlich viele Folgenglieder kommen beliebig nahe an diesen Wert.  
  Ein Häufungswert ist ein Grenzwert einer Teilfolge.

$$
a \text{ ist Häufungswert von } (a_n)
\quad\Longleftrightarrow\quad
\forall \varepsilon > 0:\ \forall N\in\mathbb{N}\ \exists n \ge N:\ |a_n - a| < \varepsilon.
$$

**Beispiel:**  
Folge $ a_n = (-1)^n + \frac{1}{n} $

- Teilfolge $ a_{2k} = 1 + \frac{1}{2k} \to 1 $
- Teilfolge $ a_{2k-1} = -1 + \frac{1}{2k-1} \to -1 $

**Häufungswerte:** $-1 $ und $ 1 $  
**Grenzwert:** keiner, da die Folge zwischen beiden Werten springt 

#### c) 

Wenn die Folge nicht gegen $a$ dann divergiert sie. Sie kann zwar bestimmt divergieren, d.h. gegen plus oder minus Undedlich oder undbestimmt wenn sie divergergiert aber nicht bestimmt. Beispeile: $\{n \mid n \in \mathbb{N}\}$ und $\{(-1)^n \mid n \in\mathbb{N}\}$

#### d)

Zu zeigen: Sei $(a_n)_{n\in\mathbb{N}}$ eine Folge reeller Zahlen, und sei $a\in\mathbb{R}$. Dann konvergiert $(a_n)_{n\in\mathbb{N}}$ genau dann gegen $a$, wenn $a$ Häufungswert jeder Teilfolge von $(a_n)_{n\in\mathbb{N}}$ ist. <i>

**Beweis des Teilfolgenprinzips:**

$\Rightarrow :$ $a_n$ konvergiert mit Grenzwert $a$. Die Indizes einer beliebigen Teilfolge $a_{n_k}$  bilden eine streng monoton wachsende Teilfolge. D.h. $\forall n_k \ge N$ gilt unsere Konvervengbedingung. 

$\Leftarrow:$ Kontraposition: $a_n$ konvergiert nicht gegen $a$ $\implies $ es gibt mindestens eine Teilfolge die nicht gegen $a$ konvergiert 

$a_n$ ist divergent dann gilt, dass $|a_n - a| > \varepsilon > 0$ fuer $n \ge N$. 

Sei $a_{n_k}$ eine beliebige aber feste Teilfolge von $a_n$ mit monoton wachsenden Indixes. Konstruire diese Folge als 

$$
  |a_1 - a | > \varepsilon_1
$$

$$
  |a_2 - a | > \varepsilon_1
$$


Aus $a_n$ mit $n_k \ge N$ somit haben wir eine Teilfolge gefunden die offensichtlich nicht gegen $a$ konvergiert und unsere 
Kontraposition is bewiesen.

### LÖSUNG AUFGABE 2

```julia
using SymPy
using Plots

function a_n(n)

    if n == 1
        return -2   
    end 

    return -2 - (3 /(a_n(n - 1) - 2))

end 

x = range(1, 10)
dp = [a_n(i) for i=1:10]

plot(x,dp)
```

```julia
function b_n(n, q)
    if n == 1
        return q
    end 
    return 2 - (3 /(b_n(n - 1, q) + 2))
end 

x = 1:10

plot(x, [
    [b_n(i, -1) for i in x],
    [b_n(i, -0.5) for i in x],
    [b_n(i, 1) for i in x],
    [b_n(i, 0.5) for i in x], 
    [b_n(i, -3) for i in x],
    [b_n(i, 3) for i in x],
    [b_n(i, -5) for i in x]
])

# Es scheint so als waere der Lim vom b_n der Lim vom -a, also 1 fuer alle b_1 != -1
```
