### LÖSUNG AUFGABE 1

```julia
function fib(n)

    if n <= 2 
        return 1
    end 

    f_1 = 1
    f_2 = 1

    for i=1:n
        a = f_2 
        f_2 = f_1 + f_2 
        f_1 = a
    end

    return f_2
end 

# a
function m(n)
    return (fib(2*n + 1) / fib(2*n))
end 

for i=1:20
    println(m(i))
end 
```

Wir wollen uns im Folgenden noch mit der Cantor-Menge $\mathcal{C}\subseteq\mathbb{R}$ beschäftigen. Zunächst beschreiben wir deren Konstruktion informell: Man startet mit dem Einheitsintervall $A_0:=[0,1]$ und entfernt daraus das ,,mittlere Drittel'' (wobei die Randpunkte jeweils nicht mit entfernt werden). Das Resultat $A_1$ besteht folglich aus den Teilintervallen $\left[0,\frac{1}{3}\right]$ und $\left[\frac{2}{3},1\right]$. Nun entfernt man aus diesen beiden Teilintervallen jeweils das mittlere Drittel und fährt immer weiter so fort. Für festes, aber beliebiges $n\in\mathbb{N}_0$ besteht $A_n$ also aus $2^n$ paarweise disjunkten abgeschlossenen Teilintervallen der Länge $3^{-n}$, und $A_{n+1}$ geht aus $A_n$ hervor, indem aus jedem dieser Teilintervalle das mittlere Drittel entfernt wird. Das, was ,,übrig bleibt'', ist die Cantor-Menge. Nun mathematisch präzise: Zu $n\in\mathbb{N}_0$ sei $A_{n+1}:=\frac{1}{3}(A_n\cup(2+A_n))$ (zur Notation - für $X\subseteq\mathbb{R}$ und $a,t\in\mathbb{R}$ ist definiert $a X:=\{a x:x\in X\}$, $t+X:=\{t+x:x\in X\}$). Dann $\mathcal{C}:=\bigcap_{n=0}^\infty A_n$.

a) 

**Annahme**

Angenommen, die Cantor-Menge sei abzaehlbar. Dann existiert eine bijektive Funktion  

$$
c:\mathbb{N} \to \mathcal{C}.
$$

Nach dem ersten Schritt der Cantor-Konstruktion gilt  

$$
A_1 = [0,\tfrac13] \cup [\tfrac23,1].
$$

Da  

$$
c(1) \in \mathcal{C} \subset A_1,
$$
liegt $c(1)$ in genau einem der beiden disjunkten Intervalle.  
Wir wählen das jeweils andere Intervall und nennen es $I_1$.

Damit gilt:

- $I_1\subset A_1$,
- $I_1$ ist abgeschlossen,
- $c(1)\notin I_1$.


**Induktionsschritt**

Angenommen, es existiert ein abgeschlossenes Intervall  

$$
I_n \subset A_n \quad\text{mit}\quad c(n)\notin I_n.
$$

Beim Uebergang von $A_n$ zu $A_{n+1}$ wird das mittlere Drittel jedes Intervalls entfernt.  
Daher zerfällt auch $I_n$ in genau zwei disjunkte abgeschlossene Teilintervalle von $A_{n+1}$.

Da $c(n+1)$ nur in einem dieser beiden Intervalle liegen kann, waehlen wir das andere und nennen es $I_{n+1}$.

Damit gilt:

- $I_{n+1}\subsetneq I_n$,
- $I_{n+1}$ ist abgeschlossen und nicht leer,
- $c(n+1)\notin I_{n+1}$


Durch die Konstruktion entsteht eine geschachtelte Folge
$$
I_1 \supsetneq I_2 \supsetneq I_3 \supsetneq \cdots
$$

Die Laengen gehen also gegen 0.

Da die Intervalle abgeschlossen, beschraenkt und geschachtelt sind und ihre Laenge gegen 0 geht, folgt aus dem Intervallschachtelungsprinzip:

$$
\bigcap_{n=1}^{\infty} I_n \neq \varnothing.
$$

Es existiert genau ein Punkt  
$$
x \in \bigcap_{n=1}^{\infty} I_n.
$$

Da jedes $I_n \subset A_n$ gilt, folgt  

$$
x \in \mathcal{C}.
$$


**Widerspruch**

Weil $c$ nach Annahme surjektiv ist, existiert ein $k\in\mathbb{N}$ mit  

$$
x = c(k).
$$

Nach unserer Konstruktion wurde jedoch $I_k$ so gewaehlt, dass  
$$
c(k)\notin I_k,
$$

waehrend  
$$
x \in I_k.
$$

Das ist ein Widerspruch.


Die Annahme, dass die Cantor-Menge abzählbar ist, führt zu einem Widerspruch.

b) Die Cantor Menge ist eine Teilmenge der Reelle Zahlen und unabzaehlbar von daher ist sind auch die reelle Zahlen unabzaehlbar
Angenommen es existierte eine Bijektion vom N auf R, dann gabe es eine Bijektion auch auf die Cantor Menge dessen Existenz schon widerlegt wurde, ergo R ist unazaelbar.
