## √Ñquivalenzrelationen


[üéûÔ∏è Relationen](https://youtu.be/BKR3z-hak2M)

Mit Relationen kann man mathematisch ausdr√ºcken, dass gewisse Elemente zueinander in Relation stehen. M√∂chten wir z.B. die Relation betrachten, dass eine Person j√ºnger als eine andere ist, dann w√ºrden wir dies mathematisch wie folgt modellieren: $M$ ist die Menge aller Menschen, und in der Relation $R$ sammeln wir nun alle Paare $(A,B)$ wobei Person $A$ j√ºnger als Person $B$ ist. 

<b> Definition </b> <br> Sei $M$ eine Menge. Eine <b> Relation </b> auf $M$ ist eine Teilmenge $R \subseteq M\times M$. Statt $(m,n) \in R$ f√ºr $m,n \in M$ schreiben wir oft auch $mRn$ und sagen "$m$ steht in Relation zu $n$". Eine Relation hei√üt... 
    
- <b>reflexiv</b>, wenn f√ºr alle $m\in M$ gilt: $mRm$, <br> d.h. alle Elemente stehen zu sich selber in Relation. 
- <b>symmetrisch</b>, wenn f√ºr alle $m,n \in M$ gilt: $mRn$ impliziert $nRm$, <br> d.h. wenn $m$ mit $n$ in Relation steht, dann muss auch $n$ zu $m$ in Relation stehen.
  
- <b>transitiv </b>, wenn f√ºr alle $m,n,l \in M$ gilt: $mRn$ und $nRl$ impliziert $mRl$, <br> d.h. wenn $m$ zu $n$ und $n$ zu $l$ in Relation stehen, dann muss auch $m$ zu $l$ in Relation stehen.


Unsere Relation im Beispiel oben ist weder reflexiv (die Menschen sind nicht j√ºnger als sie selbst) noch symmetrisch (ist $A$ j√ºnger als $B$, dann ist nicht $B$ j√ºnger als $A$) aber transitiv (ist $A$ j√ºnger als $B$ und $B$ ist j√ºnger als $C$, dann ist erst recht $A$ j√ºnger als $C$).

## √Ñquivalenzklassen und Partitionen




[üéûÔ∏è √Ñquivalenzrelationen und Partitionen](https://youtu.be/8BC5PxG7Q5Y)

Wir betrachten wieder als Beispiel auf der Menge $\mathbb{N}$ die √Ñqivalenzrelation mit $x\sim y$, wenn $x$ und $y$ die gleiche Parit√§t haben, d.h. $x$ und $y$ sind entweder beide gerade oder beide ungerade. Dann besteht die Menge aller √Ñquivalenzklassen $\{[1],[2]\}$ also aus zwei Elementen. Da die
 √Ñquivalenzklassen selber etwas sperrig sind, ist es in Anwendungen manchmal hilfreich eine Vertretermenge auszuw√§hlen, d.h. aus jeder √Ñquivalenzklasse wird ein Vertreter ausgew√§hlt.  Definition </b> <br> Sei $M$ eine Menge und $\sim$ eine √Ñquivalenzrelation auf $M$. Eine Teilmenge $V\subseteq M$ hei√üt <b> Vertretermenge </b> oder <b> Transversale </b>, wenn $V$ aus jeder √Ñquivalenzklasse genau ein Element enth√§lt.  

Ein wichtiges Beispiel einer √Ñquivalenzrelation sind die "kongruent modulo..." Relationen, wir betrachten als Beispiel die Relation $\equiv_7$ auf $\mathbb{Z}$ (Kongruenz modulo $7$). Zwei Zahlen $x,y \in \mathbb{Z}$ hei√üen kongruent modulo $7$, wenn sich beim Teilen durch $7$ der gleiche Rest ergibt. In anderen Worten: $x\equiv_7 y$ genau dann, wenn $(x-y)$ durch $7$ teilbar ist. Dies ist eine √Ñquivalenzrelation auf $\mathbb{Z}$ und wir k√∂nnen als Vertretermenge $\{0,1,2,3,4,5,6\}$ w√§hlen. Die Zahl $3$ beispielsweise vertritt dann folgende Klasse:

$$[3]=\{3,10,17,24,....\} \cup \{...,-18,-11,-4\}$$

Jede ganze Zahl hat bei Division durch $7$ einen Rest in $\{0,1,2,3,4,5,6\}$. Ist der Rest beim Teilen einer Zahl durch $7$ beispielsweise $5$, so liegt die Zahl in der Klasse $[5]$. Wir k√∂nnen $\mathbb{Z}$ also schreiben als Vereinigung 

$$\mathbb{Z}=[0]\cup [1] \cup [2] \cup [3] \cup [4] \cup [5] \cup [6] $$ 

der √Ñquivalenzklassen, und haben die ganzen Zahlen sozusagen umsortiert nach ihrem Rest beim Teilen durch $7$.

## L√∂sung Aufgabe 1



```julia
function num_transversals(m::Integer)
    
    q = div(m, 10) # volle 10ner Boxen
    r = m % 10     # restliche Elemente

    if q == 0           # wenn es keine vollen 10ner Boxen gibt
        return BigInt(1) # nur eine M√∂glichkeit
    else
        return BigInt(q)^(10 - r) * BigInt(q + 1)^r # q: klassen mit q elemente, q+1: klassen mit q+1 elemente, 
        #                                           # hoechstes potenziert mit anzahl der klassen
    end
end

function num_transversals_slow(m::Integer)
    
    classes = Dict{Int, Vector{Int}}()

    for d in 0:9
        classes[d] = Int[]
    end

    for a in 1:m
        d = a % 10      
        push!(classes[d], a)
    end

    tr = BigInt(1)

    for d in 0:9
        cls_size = length(classes[d])

        if cls_size == 0
            continue
        end

        tr *= BigInt(cls_size)
    end

    return tr
end

println(num_transversals(25))
println(num_transversals_slow(25))
```

## Wohldefiniertheit

Wenn von einer "wohldefinierten Abbildung" gesprochen wird, ist an sich nichts anderes als eine Abbildung gemeint. Es handelt sich also um eine Tautologie, die aber sinnvoll ist, um zu unterstreichen, dass es sich bei der vorliegenden Zuordnung wirklich um eine Abbildung handelt, d.h. dass **jedem Element genau ein Element im Wertebereich zugeordnet wird**.

Wir benutzen den Begriff "wohldefinierte Abbildung" dann, wenn es nicht offensichtlich ist, dass es sich um eine Abbildung handelt. Es k√∂nnte also entweder sein, dass es nicht offensichtlich ist, dass auf Werte abgebildet wird, die auch wirklich im Wertebereich liegen, oder (h√§ufiger) geht es darum, zu betonen, dass eine <b> eindeutige </b> Zuordnung vorliegt. 

Ein typischer solcher Fall ist der folgende: Sei $M$ eine Menge mit einer √Ñquivalenzrelation $\sim$ und $\mathcal{M}$ sei die Menge der √Ñquivalenzklassen und sei $N$ eine weitere Menge. Um eine Abbildung $\mathcal{M} \to N$ zu definieren, muss erkl√§rt werden, worauf die √Ñquivalenzklassen $[a]$ f√ºr $a\in M$ abgebildet werden. Oft geschieht dies dann in Abh√§ngigkeit des Vertreters $a$ und dann muss die Wohldefiniertheit gezeigt werden, d.h. es muss gezeigt werden, dass die Klasse auf den selben Wert abgebildet werden w√ºrde, wenn man einen anderen Verteter $b$ aus derselben Klasse ausw√§hlt (denn sonst w√ºrde die Abbildung dem einen Element $[a]=[b] \in \mathcal{M}$ zwei verschiedene Werte zuordnen). 




Als <u> <b> Beispiel </b> </u> betrachten wir $\mathbb{R}$ mit der √Ñquivalenzrelation $a\sim b$ genau dann wenn $|a|=|b|$, d.h. es gibt die √Ñquivalenzklassen $\{0\}$ und $\{\pm a\}$ f√ºr $a>0$. 

Folgende Vorschrift ist dann <b> keine </b> wohldefinierte Abbildung:

$$ 
f \colon \mathcal{M}\to \mathcal{M}, [a] \mapsto [a-1] 
$$ 

denn $[1]=[-1]$ aber $f([1])=[0]\neq [-2]=f([-1])$
Betrachten wir dagegen 


$$ 
g \colon \mathcal{M}\to \mathcal{M}, [a]\mapsto [a^2-1] 
$$ 

handelt es sich um eine wohldefinierte Abbildung, da $[a^2-1]=[(-a)^2-1]$ f√ºr alle $a>0$.

## Verkn√ºpfungen

Eine Verkn√ºpfung verkn√ºpft zwei Elemente einer Menge zu einem anderen Element dieser Menge, die Addition auf den nat√ºrlichen Zahlen beispielsweise ordnet zwei Zahlen $(a,b)$ das Ergebnis $a+b$ zu. 

<div class="alert alert-block alert-warning">
    <b> Definition </b> <br> Sei $M$ eine Menge. Eine <b> Verkn√ºpfung $\circ$ </b> auf $M$ ist eine Abbildung $\circ \colon M\times M\to M$. Die Verkn√ºpfung hei√üt 
<ul>
     <li> <b>assoziativ</b>, wenn f√ºr alle $a,b,c\in M$ gilt: $a\circ(b\circ c)=(a\circ b)\circ c$ </li>
        <li> <b>kommutativ</b>, wenn f√ºr alle $a,b \in M$ gilt: $a\circ b= b \circ a$  </li>
  </ul>
</div>

Ein wichtiger Punkt in der Definition von Verkn√ºpfungen ist, dass nach $M$ abgebildet wird, d.h. es muss $a\circ b \in M$ gelten f√ºr alle $a,b \in M$. Zum Beispiel definiert die √ºbliche Addition nat√ºrlicher Zahlen <b> keine </b> Addition auf $M=\{1,2,3,4,5\}$, da beispielsweise $3+4=7 \notin M$.

Typische Beispiele von Verkn√ºpfungen sind Addition und Multiplikation auf den Mengen $\mathbb{N}, \mathbb{Z}, \mathbb{Q}, \mathbb{R}$: 
$$+\colon \mathbb{N}\times \mathbb{N}\to \mathbb{N}$$
$$+\colon \mathbb{Z}\times \mathbb{Z}\to \mathbb{Z}$$
$$+\colon \mathbb{Q}\times \mathbb{Q}\to \mathbb{Q}$$
$$+\colon \mathbb{R}\times \mathbb{R}\to \mathbb{R}$$
und 
$$\cdot \colon \mathbb{N}\times \mathbb{N}\to \mathbb{N}$$
$$\cdot\colon \mathbb{Z}\times \mathbb{Z}\to \mathbb{Z}$$
$$\cdot\colon \mathbb{Q}\times \mathbb{Q}\to \mathbb{Q}$$
$$\cdot\colon \mathbb{R}\times \mathbb{R}\to \mathbb{R}$$
Diese Verkn√ºpfungen sind alle assoziativ und kommutativ. 

Wenn wir in Julia symbolisch rechnen, geht Julia √ºbrigens auch davon aus, dass Addition und Multiplikation assoziativ und kommutativ sind, d.h. dies ist in SymPy voreingestellt:

```julia
using SymPy;
a,b,c=symbols("a,b,c");
```

```julia
(a+b)+c==a+(b+c)
```

```julia
(a*b)*c==a*(b*c)
```

```julia
a+b==b+a
```

```julia
a*b==b*a
```

Ein Beispiel einer Verkn√ºpfung, die weder assoziativ noch kommutativ ist, ist die Subtraktion $-\colon \mathbb{Z}\times \mathbb{Z}\to \mathbb{Z}$, denn beispielsweise ist $(2-1)-1\neq 2-(1-1)$ und $2-1\neq 1-2$. Dies ist auch Julia bekannt:

```julia
a-(b-c)==(a-b)-c
```

```julia
a-b==b-a
```

Etwas komplizierter wird es, wenn wir Verkn√ºpfungen $M\times M \to M$ betrachten und die Elemente in $M$ selber Abbildungen sind. Hier gibt es zwei verschiedene Typen von Verkn√ºpfungen:
